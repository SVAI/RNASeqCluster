```python
import os
dataDir = "data"
filename = "fake.csv"
filepath = os.path.join(dataDir, filename)
```

# Load Data

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split

testingSplit = .2

data = pd.read_csv(filepath)
categorical = pd.get_dummies(data["label"].values).values
for i, row in data.iterrows():
    onehot = categorical[i]
    data.set_value(i, 'label', onehot)

data
```

# Data Prep

```python
data.columns
data = data.drop('Sequence Name', 1)

train, validation = train_test_split(data, test_size=testingSplit)
trainX, trainY = train[train.columns[:-1]], train["label"]
testX, testY = validation[validation.columns[:-1]], validation["label"]
trainX
```

```python
trainY.values
```

# Dimensionality Reduction

```python
pca = PCA(n_components=2)
reducedX = pca.fit_transform(trainX)
reducedTestX = pca.fit_transform(testX)
```

```python
reducedX
```

# Decision Tree

```python
from sklearn import tree
def decisionTree(trainX, trainY, validX, validY):
    clf = tree.DecisionTreeClassifier()
    clf = clf.fit(reducedX, trainY)
    prediction = clf.predict(validX)
    zOneLoss = (validY - prediction).apply(lambda x: 0 if x is 0 else 1)
    score = 1 - zOneLoss.sum()/len(zOneLoss)
    print(score)

print("Before Reduction")
decisionTree(trainX, trainY, reducedTestX, testY)
print("With Reduction")
decisionTree(reducedX, trainY, reducedTestX, testY)
```

# Logistic Regression

```python
from sklearn import linear_model
def logReg(trainX, trainY, validX, validY, regularizer_coefficient=None):
    if regularizer_coefficient is not None:
        lr = linear_model.LogisticRegression(C=regularizer_coefficient)
    else:
        lr = linear_model.LogisticRegression()
    lr.fit(trainX, trainY)
    print(lr.score(validX, validY))

print("Before Reduction")
logReg(trainX, trainY, reducedTestX, testY)
print("With Reduction")
logReg(reducedX, trainY, reducedTestX, testY)
```

# K Nearest Neighbors

```python
from sklearn import neighbors
def kNN(trainX, trainY, validX, validY, num_neighbors=3):
    nn = neighbors.KNeighborsClassifier(num_neighbors)
    nn.fit(trainX, trainY)
    print(nn.score(validX, validY))

print("Before Reduction")
kNN(trainX, trainY, testX, testY)

print("With Reduction")
kNN(reducedX, trainY, reducedTestX, testY)
```

# SVM

```python
from sklearn.svm import SVC
def svm(trainX, trainY, validX, validY, regularizer_coefficient=None):
    if regularizer_coefficient is not None:
        lr = SVC(C=regularizer_coefficient)
    else:
        lr = SVC()
    lr.fit(trainX, trainY)
    print(lr.score(validX, validY))

print("Before Reduction")
svm(trainX, trainY, testX, testY)

print("Without Reduction")
svm(reducedX, trainY, reducedTestX, testY)
```
